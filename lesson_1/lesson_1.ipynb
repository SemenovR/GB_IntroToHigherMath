### 4. Задание
Постройте на одном графике две кривые y(x) для функции двух переменной
y(k,x)=cos(k·x), взяв для одной кривой значение k=1, а для другой – любое другое k, не
равное 1.

#### Ответ:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 200)

plt.plot(x, np.cos(1*x), label="k=1")
plt.plot(x, np.cos(2.5*x), label="k=2.5")

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

## 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:<br>
1. окружности,<br>
2. эллипса,<br>
3. гиперболы.

## Решение

```python
import matplotlib.pyplot as plt
```

```python
def show_plot(x, y, title):
    plt.plot(x, y, color='blue')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(y=0, color='black')
    plt.axvline(x=0, color='black')
    plt.axis('equal')
    plt.title(title)
    plt.show()
```

#### Окружность

```python
def show_circle(x0, y0, r):
    t = np.linspace(0, 2 * np.pi, 1000)
    x = r * np.cos(t) + x0
    y = r * np.sin(t) + y0
    show_plot(x, y, 'Окружность')
```

```python
show_circle(-5, 3, 10)
```

#### Эллипс

```python
def show_ellipse(x0, y0, a, b):
    t = np.linspace(0, 2 * np.pi, 1000)
    x = a * np.cos(t) + x0
    y = b * np.sin(t) + y0
    show_plot(x, y, 'Эллипс')
```

```python
show_ellipse(2, 1, 3, 1.5)
```

#### Гипербола

```python
def show_hyperbola(x0, y0, a, b):
    t = np.linspace(-np.pi/2, np.pi/2, 1000)
    x = a * np.cosh(t) + x0
    y = b * np.sinh(t) + y0
    plt.plot(x, y, color='blue')
    x = - a * np.cosh(t) + x0
    show_plot(x, y, 'Гипербола')
```

```python
show_hyperbola(2, 1.5, 1, 1.2)
```

## 4. Задание (в программе)
1) Решите систему уравнений:
$$y = x^2 – 1$$
$$exp(x) + x·(1 – y) = 1$$

2) Решите систему уравнений и неравенств:
$$y = x^2 – 1$$
$$exp(x) + x·(1 – y) > 1$$

## Решение

#### 1. Система уравнений

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5, 5, 100)
y = x**2 - 1.
plt.plot(x, y, color='red')

x1 = np.linspace(-5, -0.01, 50)
y1 = (- 1. + np.exp(x1) + x1)/x1
plt.plot(x1, y1, color='blue')

x2 = np.linspace(0.01, 5, 50)
y2 = (- 1. + np.exp(x2) + x2)/x2
plt.plot(x2, y2, color='blue')

plt.xlabel('x')
plt.ylabel('y')
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
plt.show()
```

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1., np.exp(x) + x * (1. - y) - 1.)

res = []
res.append(fsolve(equations, (-2, 10)))
res.append(fsolve(equations, (0, 0)))
res.append(fsolve(equations, (2, 10)))
res.append(fsolve(equations, (4, 10)))

for r in res:
    print(r)
```

#### 2. Система уравнения и неравенства

Корни мы нашли на предыдущем задании. Теперь определим какой из промежутков относится к неравенству.<br>
Для этого проверим середину каждого промежутка на удовлетворение неравенству.

```python
temp_x = []
for r in res:
    temp_x.append(r[0])
temp_x.append(sys.float_info.max) #+бесконечность
temp_x.sort()
    
interval = []
x1 = sys.float_info.min #-бесконечность
for tx in temp_x:
    if tx == sys.float_info.max: #обход переполнения
        x2 = x1 + 1
    else:
        x2 = tx
    xs = (x2 + x1) / 2.
    if np.exp(xs) + xs * (1. - (xs**2 - 1.)) > 1.:
        interval.append([x1, tx])
    x1 = tx

print(interval)
```

Ответ: (0, 2.61814557) $\cup$ (4.20010584, +$\infty$)

# 4. Задание
Решите аналитически и потом численно (в программе) уравнение, зависящее от параметра а:<br>
sin(а*x)=0<br>
при условии: 0.01<a<0.02, 100<х<500.<br>
Т.е. надо найти решение х как функцию параметра а - построить график x=x(а).<br>
Если численным методом не получается найти все ветви решения x(а), то отыщите хотя бы одну.

# Решение

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
a = np.linspace(0.01, 0.02, 100)
x = np.linspace(100., 500., 100)
```

```python
def f(p):
    a, x = p
    return (np.sin(a * x), 0)
```

```python
res = []
for ai in a:
    for xi in x:
        val, info, ier, msg = fsolve(f, (ai, xi), full_output=True)
        if ier == 1 and 0.01 < val[0] < 0.02 and 100. < val[1] < 500.:
            res.append(np.around((val[0] * val[1] / np.pi)))
k = np.unique(np.array(res))
print(k)
```

```python
for ki in k:
    plt.plot(a, ki * np.pi / a)
plt.xlabel('a')
plt.ylabel('x')
plt.show()
```

```python
for ki in k:
    print(f"{ki} * Pi / a")
```
